# Visualisation of embedding space

```python
from sklearn.preprocessing import normalize
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from umap import UMAP
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

def read_embeddings(input_file, dim=1024, normalized=True):
    X = np.fromfile(input_file, dtype=np.float32, count=-1)
    X = np.resize(X, (X.shape[0] // dim, dim))
    if normalized:
        X = normalize(X)
    return X
```

```python
flores_dev_leet = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/laser/flores200/100/dev/cleaned.eng_leet_Latn.dev"
flores_dev_ugc = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/laser/flores200/100/dev/cleaned.eng_Latn_ugc.dev"
flores_dev = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/laser/flores200/100/dev/cleaned.eng_Latn.dev"
rocsmt_raw = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/laser/rocsmt/test/cleaned.src.raw-manseg.en.test"
rocsmt_norm = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/laser/rocsmt/test/cleaned.src.norm-manseg.en.test"
```

```python
flores_dev_leet_roberta = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/roberta-student/flores200/100/dev/cleaned.eng_leet_Latn.dev"
flores_dev_ugc_roberta = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/roberta-student/flores200/100/dev/cleaned.eng_Latn_ugc.dev"
flores_dev_roberta = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/roberta-student/flores200/100/dev/cleaned.eng_Latn.dev"
rocsmt_raw_roberta = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/roberta-student/rocsmt/test/cleaned.src.raw-manseg.en.test"
rocsmt_norm_roberta = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/roberta-student/rocsmt/test/cleaned.src.norm-manseg.en.test"
```

```python
flores_dev_leet_c_roberta = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/character-roberta-student/flores200/100/dev/cleaned.eng_leet_Latn.dev"
flores_dev_ugc_c_roberta = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/character-roberta-student/flores200/100/dev/cleaned.eng_Latn_ugc.dev"
flores_dev_c_roberta = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/character-roberta-student/flores200/100/dev/cleaned.eng_Latn.dev"
rocsmt_raw_c_roberta = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/character-roberta-student/rocsmt/test/cleaned.src.raw-manseg.en.test"
rocsmt_norm_c_roberta = "/home/lnishimw/scratch/experiments/noise-normalization/laser/experiment_025_lrec_xsim/embeddings/character-roberta-student/rocsmt/test/cleaned.src.norm-manseg.en.test"
```

```python
src_file = flores_dev_roberta
ref_file = flores_dev
```

```python
X_src = read_embeddings(src_file)
X_ref = read_embeddings(ref_file)
```

```python
X = np.concatenate([X_src, X_ref])
```

```python
file = np.repeat(["ugc", "std"], X_ref.shape[0])
```

```python
seed = 0
pca = PCA(n_components=3, random_state=seed)
umap = UMAP(n_components=3, init='random', random_state=seed)
tsne = TSNE(n_components=3, learning_rate='auto', init='pca', perplexity=20, random_state=seed)
```

```python
data = pd.DataFrame(columns=["PCA dim 1", "PCA dim 2", "PCA dim 3", "t-SNE dim 1", "t-SNE dim 2", "t-SNE dim 3", "UMAP dim 1", "UMAP dim 2", "UMAP dim 3", "file"])
```

```python
data["file"] = file
```

```python
data[["PCA dim 1", "PCA dim 2", "PCA dim 3"]] = pca.fit_transform(X)
data[["t-SNE dim 1", "t-SNE dim 2", "t-SNE dim 3"]] = tsne.fit_transform(X)
data[["UMAP dim 1", "UMAP dim 2", "UMAP dim 3"]] = umap.fit_transform(X)
```

```python
plt.figure(dpi=1200) 
```

```python
sns.pairplot(data[["PCA dim 1", "PCA dim 2", "PCA dim 3", "file"]], hue="file")
plt.savefig("/home/lnishimw/scratch/LASER/tasks/pca_pairplot.png")
```

```python
sns.pairplot(data[["t-SNE dim 1", "t-SNE dim 2", "t-SNE dim 3", "file"]], hue="file")
plt.savefig("/home/lnishimw/scratch/LASER/tasks/tsne_pairplot.png")
```
